## 0. Introduction

#### Qu'est-ce que l'Analyse Exploratoire des Données (EDA) ?

L'analyse exploratoire des données (EDA), désigne le processus d'examen initial d'un ensemble de données afin de comprendre ses caractéristiques, ses schémas et ses relations. L'EDA implique l'utilisation de techniques statistiques et de visualisation pour explorer les données, détecter les valeurs aberrantes, identifier les tendances, les corrélations et les motifs cachés. L'objectif principal de l'analyse exploratoire des données est de révéler des informations précieuses et de formuler des hypothèses pour orienter des analyses plus approfondies.

#### Comment effectuer une Analyse Exploratoire des Données (AED) ?

1. Collecte des données : Rassemblez toutes les données pertinentes pour votre analyse. Cela peut inclure des fichiers CSV, des bases de données, des fichiers Excel, etc.

2. Chargement des données : Importez les données dans votre environnement de 
travail Python. Utilisez des bibliothèques telles que pandas pour lire et manipuler les données.

3. Exploration initiale des données : Examinez les premières lignes de votre jeu de données à l'aide de la méthode head() de pandas. Cela vous donnera un aperçu des variables et des valeurs.

4. Gestion des valeurs manquantes : Identifiez les valeurs manquantes dans votre jeu de données et décidez de la meilleure façon de les traiter. Vous pouvez les supprimer, les remplacer par des valeurs moyennes ou médianes, ou utiliser d'autres techniques appropriées.

5. Détection des valeurs aberrantes : Recherchez les valeurs aberrantes ou les données anormales qui pourraient fausser vos résultats. Utilisez des méthodes statistiques ou des graphiques spécifiques pour repérer ces valeurs.

6. Statistiques descriptives : Calculez des statistiques descriptives telles que la moyenne, la médiane, l'écart-type, les valeurs minimales et maximales pour chaque variable. Utilisez les méthodes de pandas, telles que describe() et mean(), pour obtenir ces informations.

7. Visualisation des données : Créez des graphiques et des visualisations pour explorer les relations et les tendances entre les variables. Utilisez des bibliothèques telles que matplotlib et seaborn pour créer des histogrammes, des diagrammes en boîte, des graphiques en barres, des diagrammes de dispersion, etc.

### 6-7 Les 5 meilleures films avec la plus grande durée.

Le code extrait les films d'un DataFrame Netflix en utilisant la condition "type == 'Movie'". Le DataFrame résultant s'appelle "movies".

La colonne "duration" du DataFrame "movies" est convertie en type numérique. Les valeurs de la colonne "durée" sont modifiées en supprimant la chaîne de caractères " min" et en les convertissant en float.

Les films du DataFrame "movies" sont triés par durée dans l'ordre décroissant. Le DataFrame trié est nommé "movies_sorted".

Les cinq meilleurs films en termes de durée sont extraits du DataFrame trié en utilisant la méthode "head(5)". Le DataFrame résultant est appelé "top_5_movies".

### 6-8 Extraction des contenus similaires en comparant leurs caractéristiques textuelles.

Un vecteur de caractéristiques est créé à partir des descriptions des contenus en utilisant la classe "CountVectorizer" avec les mots vides en anglais exclus.

Les descriptions des contenus sont transformées en vecteurs de caractéristiques à l'aide de la méthode "fit_transform" du vecteur créé. Les valeurs manquantes sont remplies avec une chaîne vide.

La similarité cosinus est calculée entre chaque paire de contenus en utilisant la fonction "cosine_similarity".

L'indice du contenu "Naruto" dans le DataFrame est extrait.

La similarité cosinus entre "Naruto" et les autres contenus est récupérée et triée par ordre décroissant, en excluant "Naruto" lui-même. Les neuf contenus les plus similaires sont sélectionnés.

Les titres des contenus similaires sont affichés à l'écran en utilisant une boucle. Un message introductif est affiché avant la liste des titres des contenus similaires.

```python
#créer un vecteur de caractéristiques pour les descriptions des contenus
vector = CountVectorizer(stop_words='english')
features = vector.fit_transform(new_netflix_data['description'].fillna(''))

#calculer la similarité cosinus entre chaque paire de contenus
cosine_simil = cosine_similarity(features)

#trouver les contenus similaires à "Naruto"
index = new_netflix_data[new_netflix_data['title'] == "Naruto"].index[0]
similar_content = list(enumerate(cosine_simil[index]))
sorted_similar_content = sorted(similar_content,key=lambda x:x[1],reverse=True)[1:10]

#afficher les contenus similaires
print("Contenus similaires à Naruto :")
for content in sorted_similar_content:
    print(new_netflix_data.iloc[content[0]]['title'])
```

La similarité cosinus est une mesure utilisée pour évaluer la similarité entre deux vecteurs. Elle est souvent utilisée dans le domaine du traitement automatique du langage naturel pour comparer la similarité entre des documents textuels.

La similarité cosinus mesure la similarité directionnelle entre deux vecteurs en prenant en compte l'angle entre eux dans un espace vectoriel. Plus les vecteurs sont similaires, plus l'angle entre eux est petit, et donc la similarité cosinus est proche de 1. À l'inverse, si les vecteurs sont très différents, l'angle entre eux est grand et la similarité cosinus est proche de 0.
